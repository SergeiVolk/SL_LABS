## 1. Контейнер vs Итерируемый объект vs Итератор vs Генератор

### 1.1 Контейнеры

**Контейнер** - объект, который содержит другие объекты (списки, кортежи, словари, множества).

```python
# Примеры контейнеров
list_container = [1, 2, 3, 4, 5]
tuple_container = (1, 2, 3)
dict_container = {'a': 1, 'b': 2}
set_container = {1, 2, 3, 4}

print(f"Список: {list_container}")
print(f"Кортеж: {tuple_container}")
print(f"Словарь: {dict_container}")
print(f"Множество: {set_container}")
```

### 1.2 Итерируемые объекты

**Итерируемый объект** - объект, который может возвращать элементы по одному.

```python
# Проверка итерируемости
def is_iterable(obj):
    try:
        iter(obj)
        return True
    except TypeError:
        return False

print(f"Список итерируемый: {is_iterable([1, 2, 3])}")
print(f"Число итерируемое: {is_iterable(5)}")
print(f"Строка итерируемая: {is_iterable('hello')}")

# Итерация по итерируемому объекту
for char in "Python":
    print(char, end=' ')
```

### 1.3 Итераторы

**Итератор** - объект, который возвращает следующий элемент при вызове `next()`.

```python
# Создание и использование итератора
my_list = [1, 2, 3]
my_iterator = iter(my_list)

print(f"Первый элемент: {next(my_iterator)}")
print(f"Второй элемент: {next(my_iterator)}")
print(f"Третий элемент: {next(my_iterator)}")

# Попытка получить следующий элемент вызовет StopIteration
try:
    print(f"Четвертый элемент: {next(my_iterator)}")
except StopIteration:
    print("Итератор исчерпан!")
```

### 1.4 Функции enumerate, zip, sorted, reversed

```python
# enumerate - возвращает пары (индекс, элемент)
fruits = ['apple', 'banana', 'cherry']
print("enumerate:")
for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# zip - объединяет несколько итерируемых объектов
names = ['Emma', 'John', 'Michael']
ages = [25, 30, 35]
print("\nzip:")
for name, age in zip(names, ages):
    print(f"{name}: {age} лет")

# sorted vs sort
numbers = [3, 1, 4, 1, 5, 9, 2]
sorted_numbers = sorted(numbers)  # возвращает новый список
numbers_copy = numbers.copy()
numbers_copy.sort()  # изменяет исходный список
print(f"\nИсходный список: {numbers}")
print(f"sorted(): {sorted_numbers}")
print(f"После sort(): {numbers_copy}")

# reversed vs reverse
reversed_numbers = list(reversed(numbers))  # возвращает новый список
numbers_copy2 = numbers.copy()
numbers_copy2.reverse()  # изменяет исходный список
print(f"\nreversed(): {reversed_numbers}")
print(f"После reverse(): {numbers_copy2}")
```

### 2.3 Документирование функций с type annotations

```python
# Функция с документацией по стандарту PEP 257 и аннотациями типов
def calculate_area(radius: float) -> float:
    """
    Вычисляет площадь круга по заданному радиусу.
    
    Args:
        radius: Радиус круга. Должен быть положительным числом.
        
    Returns:
        Площадь круга, вычисленная по формуле πr².
        
    Raises:
        ValueError: Если радиус отрицательный.
        
    Examples:
        >>> calculate_area(5)
        78.53981633974483
    """
    if radius < 0:
        raise ValueError("Радиус не может быть отрицательным")
    
    import math
    return math.pi * radius ** 2

# Аннотации типов (PEP 484)
def process_data(data: list[str], reverse: bool = False) -> tuple[int, list[str]]:
    """
    Обрабатывает список строк.
    
    Args:
        data: Список строк для обработки
        reverse: Если True, возвращает список в обратном порядке
        
    Returns:
        Кортеж из количества элементов и обработанного списка
    """
    count = len(data)
    if reverse:
        data = data[::-1]
    return count, data

# Использование функций
area = calculate_area(5)
print(f"Площадь круга с радиусом 5: {area:.2f}")

count, processed = process_data(["apple", "banana", "cherry"], reverse=True)
print(f"Количество: {count}, Обработанный список: {processed}")

# Просмотр аннотаций
print(f"\nАннотации calculate_area: {calculate_area.__annotations__}")
```

**Задача 2.2:** Напишите функцию с документацией, которая проверяет палиндром

```python
# Решение задачи 2.2
def revers(pal):
    """
    Проверяет, является ли строка палиндромом.
    
    Палиндром - это строка, которая читается одинаково 
    в обоих направлениях (слева направо и справа налево).
    
    Args:
        pal (str): Строка для проверки на палиндром
        
    Returns:
        bool: True если строка является палиндромом, False в противном случае
        
    Examples:
        >>> revers("aba")
        True
        >>> revers("abc")
        False
        >>> revers("Aba")
        False
    """
    return pal == pal[::-1]

a = "aba"
print(revers(a))  
```

**Задача 2.3:** Создайте функцию с аннотациями типов для вычисления среднего значения

```python
# Решение задачи 2.3
def middle(list1):
    """
    Вычисляет среднее арифметическое элементов списка целых чисел.
    
    Args:
        list1: Список целых чисел
        
    Returns:
        Среднее арифметическое как число с плавающей точкой
    """
    a = sum(list1)
    b = len(list1)
    return a / b

a = [1, 2, 3, 4, 8]
print(middle(a))
```

**Задача 2.5:** Создайте функцию поиска элемента в списке с полной документацией по PEP 257

```python
# Решение задачи 2.5
def finds(list1, find1):
    """
    Проверяет наличие элемента в списке.
    
    Функция выполняет линейный поиск элемента в списке и возвращает True,
    если элемент найден, и False в противном случае.

    Args:
        list1 (list): Список, в котором выполняется поиск элемента.
        find1 (any): Элемент, который необходимо найти в списке.
                    Может быть любого типа, поддерживающего сравнение.

    Returns:
        bool: True если элемент найден в списке, False в противном случае.

    Examples:
        >>> finds([1, 2, 3, 4, 5], 3)
        True
        
        >>> finds(['apple', 'banana', 'orange'], 'grape')
        False
        
        >>> finds([1, 5, 7, 9, "name"], "name")
        True

    Notes:
        - Функция использует линейный поиск, временная сложность O(n)
        - Для больших списков рассмотрите использование множеств (set)
        - Функция возвращает результат после первого найденного совпадения
    """
    for x in list1:
        if x == find1:
            return True
    else:
        return False        


a = [1, 5, 7, 9, "name"]
f = "name"
print(finds(a, f))
```

### 3.2 Особенности аргументов по умолчанию

```python
# Проблема с изменяемыми объектами по умолчанию
def add_to_list_bad(item, my_list=[]):  # ОПАСНО!
    """Неправильная реализация - список создается один раз!"""
    my_list.append(item)
    return my_list

def add_to_list_good(item, my_list=None):
    """Правильная реализация - создаем новый список при каждом вызове."""
    if my_list is None:
        my_list = []
    my_list.append(item)
    return my_list

# Демонстрация проблемы
print("Неправильная реализация:")
list1 = add_to_list_bad(1)
print(f"Первый вызов: {list1}")
list2 = add_to_list_bad(2)
print(f"Второй вызов: {list2}")  # Ожидается [2], но получаем [1, 2]!

print("\nПравильная реализация:")
list3 = add_to_list_good(1)
print(f"Первый вызов: {list3}")
list4 = add_to_list_good(2)
print(f"Второй вызов: {list4}")  # Теперь правильно: [2]
```

**Задача 3.3:** Создайте функцию `format_text` для форматирования текста с аргументами по умолчанию: выравнивание (по умолчанию 'left'), ширина (по умолчанию 20), символ-заполнитель (по умолчанию ' ').

Примеры вызова:
- `format_text("текст")` → "текст              "
- `format_text("текст", width=10)` → "текст      "
- `format_text("текст", align='center', fill_char='-')` → "-------текст-------"

```python
# Решение задачи 3.3
def format_text(text, align='left', width=20, fill_char=' '):
    if align == 'left':
        return text.ljust(width, fill_char)
    elif align == 'right':
        return text.rjust(width, fill_char)
    elif align == 'center':
        return text.center(width, fill_char)

print(format_text("текст"))
print(format_text("текст", width=10))
print(format_text("текст", align='center', fill_char='-'))
```

```python
# Решение задачи 3.5
def filter_data(data, filter_even=True, filter_positive=True):
    result = []
    for num in data:
        if filter_even and num % 2 != 0:
            continue
        if filter_positive and num <= 0:
            continue 
        
        result.append(num)
    
    return result

a = [-2, -4, 0, 1, 4, 5, 9, 7, -2]

print(filter_data(a))
print(filter_data(a, True, False))
```

### 4.2 Ключевые аргументы

```python
# Функция с ключевыми аргументами
def send_email(to, subject, body, cc=None, bcc=None, reply_to=None, priority="normal"):
    """
    Отправляет электронное письмо.
    
    Args:
        to: Адрес получателя
        subject: Тема письма
        body: Текст письма
        cc: Копия
        bcc: Скрытая копия
        reply_to: Адрес для ответа
        priority: Приоритет письма
    """
    email_info = f"To: {to}\nSubject: {subject}\nBody: {body}\nPriority: {priority}"
    
    if cc:
        email_info += f"\nCC: {cc}"
    if bcc:
        email_info += f"\nBCC: {bcc}"
    if reply_to:
        email_info += f"\nReply-To: {reply_to}"
    
    return email_info

# Использование ключевых аргументов
email2 = send_email(
    to="user@example.com",
    subject="Важное сообщение",
    body="Пожалуйста, ответьте ASAP",
    cc="manager@example.com",
    reply_to="support@example.com",
    priority="high"
)
print("Письмо с дополнительными параметрами:")
print(email2)
```

## 8. Пользовательские генераторы

### 8.1 Генераторы с yield

```python
# Простой генератор
def countdown(n):
    """Генератор обратного отсчета."""
    print("Начало отсчета")
    while n > 0:
        yield n
        n -= 1
    print("Отсчет завершен!")

# Использование генератора
print("Обратный отсчет:")
for number in countdown(5):
    print(number)

# Генератор можно использовать вручную
print("\nРучное использование:")
counter = countdown(3)
print(next(counter))
print(next(counter))
print(next(counter))

# После завершения генератор вызывает StopIteration
try:
    print(next(counter))
except StopIteration:
    print("Генератор завершен")
```

### 8.2 Преимущества генераторов

```python
# Сравнение подхода с списком и генератором
def squares_list(n):
    """Возвращает список квадратов."""
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result

def squares_generator(n):
    """Генератор квадратов."""
    for i in range(n):
        yield i ** 2

# Для больших n генератор экономит память
n = 1000000

print("Создание списка квадратов...")
# squares = squares_list(n)  # Занимает много памяти
print("Список создан (закомментировано для экономии памяти)")

print("Создание генератора квадратов...")
squares_gen = squares_generator(n)
print("Генератор создан")

# Получение первых 5 значений из генератора
print("Первые 5 квадратов:")
for i, square in enumerate(squares_gen):
    if i >= 5:
        break
    print(square)

print("Генератор продолжает работать с того места, где остановился...")
```

## 9. ООП. Классы, объекты, поля, методы, атрибуты

### 9.1 Основы ООП в Python

```python
# Простой класс
class Dog:
    """Простой класс, представляющий собаку."""
    
    # Атрибут класса (общий для всех экземпляров)
    species = "Canis familiaris"
    
    def __init__(self, name, age):
        """Конструктор класса.
        
        Args:
            name (str): Имя собаки
            age (int): Возраст собаки
        """
        # Атрибуты экземпляра (уникальные для каждого объекта)
        self.name = name
        self.age = age
    
    # Метод экземпляра
    def bark(self):
        """Собака лает."""
        return f"{self.name} говорит: Гав!"
    
    def get_info(self):
        """Возвращает информацию о собаке."""
        return f"{self.name}, {self.age} лет, вид: {self.species}"

# Создание объектов (экземпляров класса)
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

# Доступ к атрибутам
print(f"dog1.name: {dog1.name}")
print(f"dog2.name: {dog2.name}")
print(f"dog1.species: {dog1.species}")
print(f"dog2.species: {dog2.species}")

# Вызов методов
print(dog1.bark())
print(dog2.bark())
print(dog1.get_info())
print(dog2.get_info())

# Изменение атрибутов
dog1.age = 4
print(f"После изменения: {dog1.get_info()}")
```

### 9.2 Специальные методы

```python
class Book:
    """Класс, представляющий книгу."""
    
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    # Строковое представление
    def __str__(self):
        return f"'{self.title}' by {self.author}"
    
    # Формальное строковое представление
    def __repr__(self):
        return f"Book('{self.title}', '{self.author}', {self.pages})"
    
    # Длина книги
    def __len__(self):
        return self.pages

# Использование
book1 = Book("War and Peace", "Leo Tolstoy", 1225)
book2 = Book("Crime and Punishment", "Fyodor Dostoevsky", 671)

print(f"str(book1): {str(book1)}")
print(f"repr(book1): {repr(book1)}")
print(f"Длина книги: {len(book1)} страниц")
```

```python
# Решение задачи 9.5
class MathUtils:
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def multiply(a, b):
        return a * b
    
    @classmethod
    def create_from_string(cls, string_data):
        numbers = [float(x) for x in string_data.split()]
        return cls(*numbers)

print(MathUtils.add(5, 3))
print(MathUtils.multiply(4, 6))

obj = MathUtils.create_from_string("10 20")
print(obj)
print(obj.add(5, 2))
print(obj.multiply(3, 4))
```

### 10.1 Инкапсуляция

```python
class BankAccount:
    """Класс банковского счета с инкапсуляцией."""
    
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder
        self._balance = initial_balance  # Защищенный атрибут
        self.__account_id = hash(account_holder)  # Приватный атрибут
    
    # Публичные методы для взаимодействия
    def deposit(self, amount):
        """Пополнение счета."""
        if amount > 0:
            self._balance += amount
            return True
        return False
    
    def withdraw(self, amount):
        """Снятие со счета."""
        if 0 < amount <= self._balance:
            self._balance -= amount
            return True
        return False
    
    def get_balance(self):
        """Получение баланса."""
        return self._balance

# Использование
account = BankAccount("Alex Johnson", 1000)

print(f"Владелец: {account.account_holder}")
print(f"Баланс: {account.get_balance()}")

account.deposit(500)
print(f"После пополнения: {account.get_balance()}")

account.withdraw(200)
print(f"После снятия: {account.get_balance()}")

# Попытка доступа к защищенным и приватным атрибутам
print(f"\nЗащищенный атрибут (_balance): {account._balance}")  # Доступно, но не рекомендуется
# print(f"Приватный атрибут (__account_id): {account.__account_id}")  # Ошибка!
```

### 10.2 Наследование

```python
# Базовый класс
class Animal:
    """Базовый класс для всех животных."""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def speak(self):
        """Животное издает звук."""
        raise NotImplementedError("Подклассы должны реализовать этот метод")
    
    def get_info(self):
        """Возвращает информацию о животном."""
        return f"{self.name}, {self.age} лет"

# Производные классы
class Dog(Animal):
    """Класс собаки."""
    
    def __init__(self, name, age, breed):
        super().__init__(name, age)  # Вызов конструктора родительского класса
        self.breed = breed
    
    def speak(self):
        """Собака лает."""
        return "Гав!"
    
    def get_info(self):
        """Возвращает информацию о собаке."""
        return f"{super().get_info()}, порода: {self.breed}"

class Cat(Animal):
    """Класс кошки."""
    
    def speak(self):
        """Кошка мяукает."""
        return "Мяу!"

# Использование
animals = [
    Dog("Buddy", 3, "Labrador"),
    Cat("Whiskers", 2)
]

for animal in animals:
    print(f"{animal.get_info()} говорит: {animal.speak()}")

# Проверка типов
print(f"\nПроверка типов:")
print(f"Buddy - Animal: {isinstance(animals[0], Animal)}")
print(f"Buddy - Dog: {isinstance(animals[0], Dog)}")
print(f"Buddy - Cat: {isinstance(animals[0], Cat)}")
```

### 10.3 Абстрактные классы

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    """Абстрактный класс геометрической фигуры."""
    
    def __init__(self, color):
        self.color = color
    
    @abstractmethod
    def area(self):
        """Вычисляет площадь фигуры."""
        pass
    
    @abstractmethod
    def perimeter(self):
        """Вычисляет периметр фигуры."""
        pass
    
    def get_info(self):
        """Возвращает информацию о фигуре."""
        return f"Фигура цвета {self.color}, площадь: {self.area():.2f}, периметр: {self.perimeter():.2f}"

class Rectangle(Shape):
    """Класс прямоугольника."""
    
    def __init__(self, color, width, height):
        super().__init__(color)
        self.width = width
        self.height = height
    
    def area(self):
        """Площадь прямоугольника."""
        return self.width * self.height
    
    def perimeter(self):
        """Периметр прямоугольника."""
        return 2 * (self.width + self.height)

class Circle(Shape):
    """Класс круга."""
    
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius
    
    def area(self):
        """Площадь круга."""
        import math
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        """Периметр круга (длина окружности)."""
        import math
        return 2 * math.pi * self.radius

# Использование
shapes = [
    Rectangle("красный", 5, 3),
    Circle("синий", 4)
]

for shape in shapes:
    print(shape.get_info())

# Попытка создать экземпляр абстрактного класса
try:
    shape = Shape("черный")
except TypeError as e:
    print(f"\nОшибка: {e}")
```

**Задача 10.1:** Создайте иерархию классов для сотрудников компании

```python
class Worker:
    def __init__(self, full_name, worker_id, wage):
        self.full_name = full_name
        self.worker_id = worker_id
        self.wage = wage

    def perform_work(self):
        return f"{self.full_name} выполняет свои обязанности."

    def display_info(self):
        return f"Сотрудник {self.full_name} (ID: {self.worker_id}) имеет зарплату {self.wage} руб."

    def raise_wage(self, bonus):
        self.wage = self.wage + bonus
        return self.wage

class TeamLead(Worker):
    def __init__(self, full_name, worker_id, wage, years_experience):
        Worker.__init__(self, full_name, worker_id, wage)
        self.years_experience = years_experience

    def perform_work(self):
        return f"Руководитель {self.full_name} координирует работу команды."
    
    def show_experience(self):
        return f"Руководитель {self.full_name} (ID: {self.worker_id}) работает {self.years_experience} лет"

class Programmer(Worker):
    def __init__(self, full_name, worker_id, wage, tech_stack, job_level):
        Worker.__init__(self, full_name, worker_id, wage)
        self.tech_stack = tech_stack
        self.job_level = job_level

    def perform_work(self):
        return f"Разработчик {self.full_name} (уровень: {self.job_level}) программирует на {self.tech_stack}"

class Trainee(Worker):
    def __init__(self, full_name, worker_id, wage, months_count):
        Worker.__init__(self, full_name, worker_id, wage)
        self.months_count = months_count

    def perform_work(self):
        return f"Стажёр {self.full_name} ассистирует коллегам"
    
    def get_internship_info(self):
        m = self.months_count
        
        if m > 18:
            return f"Стажировка {self.full_name} затянулась, пора переводить на постоянную должность"
        
        last_digit = m % 10
        last_two = m % 100
        
        if last_digit == 1 and last_two != 11:
            word = "месяц"
        elif 2 <= last_digit <= 4 and not (12 <= last_two <= 14):
            word = "месяца"
        else:
            word = "месяцев"
        
        return f"Стажёр {self.full_name} проходит обучение уже {m} {word}"

# Тестирование
lead = TeamLead("Даник", "M001", 100000, 1)
coder = Programmer("Сергей", "D001", 80000, "Python", "Junior Developer")
trainee = Trainee("Дима", "I001", 20000, 3)

print(lead.perform_work())
print(lead.show_experience())
print()
print(coder.perform_work())
print()
print(trainee.perform_work())
print(trainee.get_internship_info())
print()
print(f"Текущая зарплата: {coder.wage}")
updated_wage = coder.raise_wage(10000)
print(f"Новая зарплата: {updated_wage}")
```

**Задача 10.2:** Реализуйте абстрактный класс для игровых персонажей

```python
from abc import ABC, abstractmethod
import random


class GameCharacter(ABC):
    def __init__(self, char_name, hp, attack_power, char_level):
        self.char_name = char_name
        self.hp = hp
        self.attack_power = attack_power
        self.char_level = char_level

    def show_stats(self):
        return f"{self.char_name} (уровень {self.char_level}): HP={self.hp}, Атака={self.attack_power}"
    
    @abstractmethod
    def strike(self):
        pass

    @abstractmethod
    def take_hit(self, incoming_damage):
        pass

    @abstractmethod
    def upgrade(self):
        pass


class Wizard(GameCharacter):
    def __init__(self, hp, attack_power, char_level, magic_points, char_name="Волшебник"):
        GameCharacter.__init__(self, char_name, hp, attack_power, char_level)
        self.magic_points = magic_points

    def strike(self):
        spell_cost = 30
        if self.magic_points >= spell_cost:
            self.magic_points -= spell_cost
            return f"{self.char_name} использует файерболл! Урон: {self.attack_power}. Мана: {self.magic_points}"
        else:
            return f"{self.char_name} недостаточно маны!"
    
    def take_hit(self, incoming_damage):
        self.hp -= incoming_damage
        if self.hp <= 0:
            self.hp = 0
            return f"{self.char_name} пал в битве"
        return f"{self.char_name} принял {incoming_damage} урона. Осталось HP: {self.hp}"
    
    def upgrade(self):
        max_level = 15
        if self.char_level < max_level:
            self.char_level += 1
            self.hp += 50
            self.attack_power += 45
            self.magic_points += 40
            return f"{self.char_name} получил уровень! (+45 атака, +50 HP, +40 мана)"
        else:
            return f"{self.char_name} достиг максимального уровня"


class Rogue(GameCharacter):
    def __init__(self, hp, attack_power, char_level, critical_rate, char_name="Убийца"):
        GameCharacter.__init__(self, char_name, hp, attack_power, char_level)
        self.critical_rate = critical_rate
        self.strike_chain = 0

    def strike(self):
        self.strike_chain += 1
        roll = random.randint(1, 100)
        
        if roll <= self.critical_rate:
            critical_hit = self.attack_power * 2
            return f"{self.char_name} нанёс крит! Урон: {critical_hit} (серия: {self.strike_chain})"
        else:
            return f"{self.char_name} бьёт из-за угла. Урон: {self.attack_power} (серия: {self.strike_chain})"
    
    def take_hit(self, incoming_damage):
        self.strike_chain = 0
        self.hp -= incoming_damage
        if self.hp <= 0:
            self.hp = 0
            return f"{self.char_name} пал в битве"
        return f"{self.char_name} принял {incoming_damage} урона. Серия сброшена! HP: {self.hp}"
    
    def upgrade(self):
        max_level = 15
        if self.char_level < max_level:
            self.char_level += 1
            self.hp += 40
            self.attack_power += 55
            self.critical_rate += 3
            return f"{self.char_name} получил уровень! (+55 атака, +40 HP, +3% крит)"
        else:
            return f"{self.char_name} достиг максимального уровня"


class Medic(GameCharacter):
    def __init__(self, hp, attack_power, char_level, restoration, char_name="Лекарь"):
        GameCharacter.__init__(self, char_name, hp, attack_power, char_level)
        self.restoration = restoration
        self.hp_limit = hp

    def strike(self):
        return f"{self.char_name} бьёт посохом. Урон: {self.attack_power}"
    
    def restore_health(self):
        amount = min(self.restoration, self.hp_limit - self.hp)
        self.hp += amount
        return f"{self.char_name} вылечил {amount} HP! Текущее HP: {self.hp}"
    
    def take_hit(self, incoming_damage):
        self.hp -= incoming_damage
        if self.hp <= 0:
            self.hp = 0
            return f"{self.char_name} пал в битве"
        return f"{self.char_name} принял {incoming_damage} урона. HP: {self.hp}"
    
    def upgrade(self):
        max_level = 15
        if self.char_level < max_level:
            self.char_level += 1
            self.hp += 80
            self.hp_limit += 80
            self.attack_power += 20
            self.restoration += 60
            return f"{self.char_name} получил уровень! (+20 атака, +80 HP, +60 лечение)"
        else:
            return f"{self.char_name} достиг максимального уровня"


# Тестирование
wizard = Wizard(hp=600, attack_power=120, char_level=10, magic_points=150)
rogue = Rogue(hp=700, attack_power=90, char_level=8, critical_rate=25)
medic = Medic(hp=800, attack_power=50, char_level=12, restoration=100)

print(wizard.strike())
print(wizard.strike())
print(rogue.strike())
print(rogue.strike())
print(medic.strike())
print(medic.take_hit(200))
print(medic.restore_health())
```

**Задача 10.4:** Напишите класс `AdvancedCalculator`, который переопределяет методы базового класса `Calculator`. Добавьте новые методы для вычисления степени и корня.

```python
# Решение задачи 10.4
class Calculator:

    def plus(self, a, b):
        return a + b
    
    def minus(self, a, b):
        return a - b
    
    def mnoj(self, a, b):
        return a * b
    
    def delen(self, a, b):
        return a / b
    
class AdvancedCalculator(Calculator):

    def new_delen(self, a, b):
        if b == 0:
            return "Ошибка: деление на ноль"
        return a / b
    
    def power(self, num, c):
        return num ** c
    
    def sqrtn(self, num, c):  
        return num ** (1/c)  
    
calc = AdvancedCalculator()
print(f"5 + 3 = {calc.plus(5, 3)}")
print(f"10 - 4 = {calc.minus(10, 4)}")
print(f"6 * 7 = {calc.mnoj(6, 7)}")
print(f"15 / 3 = {calc.delen(15, 3)}")
print(f"8 / 0 = {calc.new_delen(8, 0)}")  
print(f"2 в степени 8 = {calc.power(2, 8)}")
print(f"5 в степени 3 = {calc.power(5, 3)}")
print(f"Корень 2-й степени из 16 = {calc.sqrtn(16, 2)}")
print(f"Корень 3-й степени из 27 = {calc.sqrtn(27, 3)}")
```
